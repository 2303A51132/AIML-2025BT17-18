4 - Data Preprocessing

```python
car_train = train_data.drop('binaryClass', axis=1)
car_labels = train_data[['binaryClass']]
```

```python
from sklearn.preprocessing import OrdinalEncoder

from sklearn.pipeline import Pipeline

pipe = Pipeline([('ord_enc', OrdinalEncoder())])

car_train_prepared = pipe.fit_transform(car_train)
```

```python
from sklearn.preprocessing import LabelEncoder

label_enc = LabelEncoder()

car_labels_prepared = label_enc.fit_transform(car_labels)
```

```python
car_train_prepared.shape, car_labels_prepared.shape
```

5 - Training Decision Tree Classifier

```python
from sklearn.tree import DecisionTreeClassifier

tree_clf = DecisionTreeClassifier()
tree_clf.set_params(class_weight='balanced', criterion='entropy')

tree_clf.fit(car_train_prepared, car_labels_prepared)
```

```python
from sklearn.tree import plot_tree

plt.figure(figsize=(20,15))
plot_tree(tree_clf, feature_names=car_train.columns,
          class_names=['N','P']);
```

```python
# Let's hide warnings returned by grid search

import warnings
warnings.filterwarnings('ignore')
```

```python
from sklearn.model_selection import GridSearchCV

params_grid = {'max_leaf_nodes': list(range(0, 10)),
               'min_samples_split': [0,1,2, 3, 4],
               'min_samples_leaf': [0,1,2, 3, 4],
               'max_depth':[0,1,2,3,4,5],
               'max_features':[0,1,2,3,4,5],
              'max_leaf_nodes':[0,1,2,3,4,5]}

#refit is true by default. The best estimator is trained on the whole dataset

grid_search = GridSearchCV(DecisionTreeClassifier(random_state=42, class_weight='balanced'), params_grid, verbose=1, cv=3, refit=True)

grid_search.fit(car_train_prepared, car_labels_prepared)
```

```python
grid_search.best_params_
```

```python
from sklearn.metrics import accuracy_score  # Import the accuracy_score function

# Assuming your accuracy calculation logic is similar to this:
def accuracy(X, model, y_true):
    """
    Calculates the accuracy of a model's predictions.

    Args:
        X: The input features.
        model: The trained model.
        y_true: The true target values.

    Returns:
        The accuracy score.
    """
    y_pred = model.predict(X)
    return accuracy_score(y_true, y_pred)

# Now you can call the accuracy function:
accuracy(car_train_prepared, tree_best, car_labels_prepared)
```

```python
from sklearn.metrics import confusion_matrix # Import the confusion_matrix function

def conf_matrix(X, model, y_true):
    """
    Generates and prints the confusion matrix.

    Args:
        X: The input features.
        model: The trained model.
        y_true: The true target values.
    """
    y_pred = model.predict(X)
    cm = confusion_matrix(y_true, y_pred)
    print(cm)
    # Optionally, you can add further visualization or analysis of the confusion matrix here

# Now you can call the conf_matrix function as before:
conf_matrix(car_train_prepared, tree_best, car_labels_prepared)
```

```python
from sklearn.metrics import classification_report # Import the classification_report function

def class_report(X, model, y_true):
    """
    Generates and prints the classification report.

    Args:
        X: The input features.
        model: The trained model.
        y_true: The true target values.
    """
    y_pred = model.predict(X)
    report = classification_report(y_true, y_pred)
    print(report)
    # Optionally, you can add further analysis or visualization of the report here

# Now you can call the class_report function:
class_report(car_train_prepared, tree_best, car_labels_prepared)
```
